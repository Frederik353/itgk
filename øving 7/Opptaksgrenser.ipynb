I denne oppgaven skal vi lese inn en fil med opptaksgrensene fra Samordna Opptak.

Filen er på CSV-format (Comma Separated Values), noe som betyr at hver linje er en liste med felter separert med komma. Tekstfelter er omsluttet av fnutter (").

* Første felt er studiets navn
* Andre felt er poenggrensen (enten et tall, eller "Alle" dersom alle kom inn)

F.eks. linjen: **"NTNU 194459 Antikkens kultur","Alle"** sier at alle som søkte kom inn på Dragvoll-studiet “Antikkens kultur” ved NTNU.

Hver funksjon i de følgende deloppgavene tar data fra filen **poenggrenser_2011.csv** som input. Derfor er det veldig praktisk å lagre innholdet i en variabel, slik at du slipper å lese den på nytt hver gang.

### a)

Les fra fila `poenggrenser_2011.csv` og lagre innholdet i en variabel.

***Skriv koden din i boksen under.***

````python
import csv

with open('poenggrenser_2011.csv', 'r') as file:
    reader = csv.reader(file)
    poenggrenser = list(reader)

for skole in poenggrenser:
    print(skole)
````

### b)

Skriv en funksjon som finner ut hvor mange studier som tok inn alle søkere. 

***Husk at du nå i alle deloppgavene kan bruke variabelen du definerte i a så lenge du har kjørt den kodesnutten først!***

*Eksempel på kjøring av kode:*
```python
Antall studier hvor alle kom inn: 590
```
***Skriv koden din i boksen under.***

````python
def antall_studier_alle_kom_inn(data):
    count = 0
    for row in data:
        poenggrense = row[1]
        if poenggrense == "0" or poenggrense.lower() == "alle":
            count += 1
    return count

antall = antall_studier_alle_kom_inn(poenggrenser)
print(f"Antall studier hvor alle kom inn: {antall}")
````

### c)

Skriv en funksjon som finner gjennomsnittlig opptaksgrense for NTNU. Ikke ta med studier som tok inn alle søkere.

*Eksempel på kjøring av kode:*
```python
Gjennomsnittlig opptaksgrense for NTNU var: 46.29
```
***Skriv koden din i boksen under.***

````python
def gjennomsnittlig_opptaksgrense_ntnu(data):
    total_poeng = 0
    antall_studier = 0
    
    for row in data:
        universitet = row[0]
        poenggrense = row[1]
        
        if "NTNU" in universitet and poenggrense not in ["0", "Alle kom inn"]:
            try:
                total_poeng += float(poenggrense)
                antall_studier += 1
            except ValueError:
                pass

    if antall_studier == 0:
        return 0
    return total_poeng / antall_studier

gjennomsnitt = gjennomsnittlig_opptaksgrense_ntnu(poenggrenser)
print(f"Gjennomsnittlig opptaksgrense for NTNU var: {gjennomsnitt:.2f}")
````

### d)

Skriv en funksjon som finner studiet med laveste opptaksgrense (som IKKE tok inn alle søkere).

*Eksempel på kjøring av kode:*
```python
Studiet som hadde den laveste opptaksgrensen var: AHO 189343 Industridesign
```
***Skriv koden din i boksen under.***

````python
def studie_med_lavest_opptaksgrense(data):
    laveste_poenggrense = float('inf')
    studie_navn = ""

    for row in data:
        navn = row[0]
        poenggrense = row[1]

        if poenggrense not in ["0", "Alle kom inn"]:
            try:
                poeng = float(poenggrense)
                if poeng < laveste_poenggrense:
                    laveste_poenggrense = poeng
                    studie_navn = navn
            except ValueError:
                pass

    return studie_navn

laveste_studie = studie_med_lavest_opptaksgrense(poenggrenser)
print(f"Studiet som hadde den laveste opptaksgrensen var: {laveste_studie}")
````

***Skriv koden din i boksen under.***

````python
def lag_studiested_dictionary(data):
    studiested_dict = {}
    # ['ATH 255418 Interkulturell forstÃ¥else Ã¥rsstudium', 'Alle']
    for row in data:
        studiested = row[0].split()[0]
        linje_navn = row[0].split()[2]
        poenggrense = row[1]

        if poenggrense.lower() != "alle":
            try:
                poeng = float(poenggrense)
                if studiested not in studiested_dict:
                    studiested_dict[studiested] = {}
                studiested_dict[studiested][linje_navn] = poeng
            except ValueError:
                pass
    return studiested_dict

# Bruker funksjonen med data fra CSV-filen
studiested_data = lag_studiested_dictionary(poenggrenser)



for studiested, linjer in studiested_data.items():
    print(studiested)
    for linje, poeng in linjer.items():
        print(f"  {linje}: {poeng}")
    print()

````
